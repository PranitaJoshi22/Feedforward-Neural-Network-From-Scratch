Q.1 To implement a simple Feedforward Neural Network from scratch using Python (without deep learning libraries), demonstrate forward propagation, backpropagation, and gradient descent, and analyze its learning behavior.

```python
import random
import math
import matplotlib.pyplot as plt

# Sigmoid Activation
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Dataset
inputs = [[0,0], [0,1], [1,0], [1,1]]
targets = [0, 0, 0, 1]

# Initialize weights
random.seed(1)

w1, w2 = random.uniform(-1,1), random.uniform(-1,1)
w3 = random.uniform(-1,1)
b1, b2 = random.uniform(-1,1), random.uniform(-1,1)

learning_rate = 0.5
epochs = 20
losses = []

for epoch in range(epochs):
    total_error = 0

    for i in range(len(inputs)):
        x1, x2 = inputs[i]
        y = targets[i]

        # Forward Pass
        zh = x1*w1 + x2*w2 + b1
        h = sigmoid(zh)

        zo = h*w3 + b2
        y_hat = sigmoid(zo)

        error = y - y_hat
        total_error += error**2

        # Backpropagation
        delta_o = error * sigmoid_derivative(y_hat)
        delta_h = delta_o * w3 * sigmoid_derivative(h)

        # Update Weights
        w3 += learning_rate * delta_o * h
        b2 += learning_rate * delta_o

        w1 += learning_rate * delta_h * x1
        w2 += learning_rate * delta_h * x2
        b1 += learning_rate * delta_h

    losses.append(total_error)

    print(f"\nEpoch {epoch+1}")
    print(f"w1={w1:.4f}, w2={w2:.4f}, b1={b1:.4f}")
    print(f"w3={w3:.4f}, b2={b2:.4f}")
    print(f"Loss={total_error:.4f}")
```

```python
plt.plot(losses)
plt.title("Loss vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
